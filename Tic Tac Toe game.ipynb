```python
#creates a dictionary for the positions of the board( 1 - Top right 2 - top middle etc)
#the vaue for each key is a space, ' ', which will be changed when either 'x' or 'O' is played in the position corresponding to the key
#Also a list of the keys so that once a player selects a particular position, that position can not be selected again.

import random

board = {i:' ' for i in ['1','2','3','4','5','6','7','8','9']}    
board_keys = ['1','2','3','4','5','6','7','8','9']

#This will print the game board

def print_board(board):
    
    print(board['1'] + '|' + board['2'] + '|' + board['3'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['7'] + '|' + board['8'] + '|' + board['9'])

#This function will check is a number string is in list. This is to carry our the comment in line 3 above of the list of keys

    
print('TIC TAC TOE  X0X0X0X0X0X0X0X0X0')

print('See guide to positions and their corresponding number below')
print('1|2|3')
print('-+-+-')
print('4|5|6')
print('-+-+-')
print('7|8|9')

print('Single Player(with Computer) or Double Players')
choice = input("Enter '1' for single player or '2' for double players:>>")   #This would allow the player to select the modes
while choice not in ['1','2']:   #This while loop will prevent players from entering invalid values.
    print('invalid entry')
    choice = input("Enter '1' for single player or '2' for double players.")

if choice == '1':
    print('You have opted to play with the computer')
    player1 = input ('Enter your name:>>')
    choice2 = input(f'{player1}, would you like to be "X" or "O". Enter "X" or "O":>>')  #player gets to choose to be x or o.
    while choice2.lower() not in ['x','o']:
        print('invalid entry')
        choice2 = input(f'{player1}, would you like to be "X" or "O". Enter "X" or "O":>>')  #player gets to choose to be x or o.
    count = 0
    win = True
    if choice2.upper() == 'X':
        comp = 'O'
    else:
        comp = 'X'
    turn,sym = player1,choice2.upper()
    print_board(board)
    while count != 9:
        print(f'{turn}-{sym}, your turn to play.')
        if turn == player1:
            pos = input(f'{turn},where would you like to play:>>')
            while pos not in board_keys:
                print('invalid entry')
                pos = input('where would you like to play:>>')
            board_keys.remove(pos)
            board[pos] = sym
            print_board(board)
            count += 1
            if count >= 5:
                if board['7'] == board['8'] == board['9'] == sym: # across the bottom
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['4'] == board['5'] == board['6'] == sym: # across the middle
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['1'] == board['2'] == board['3'] == sym: # across the top
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['1'] == board['4'] == board['7'] == sym: # down the left side
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['2'] == board['5'] == board['8'] == sym: # down the middle
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['3'] == board['6'] == board['9'] == sym: # down the right side
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win= False
                    break 
                elif board['7'] == board['5'] == board['3'] == sym: # diagonal
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['1'] == board['5'] == board['9'] == sym: # diagonal
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                else:
                    turn,sym = 'computer', comp
            else:
                turn,sym = 'computer', comp
        elif turn == 'computer':
            print('Computer.......')
            pos = random.choice(board_keys)
            board_keys.remove(pos)
            board[pos] = sym
            print_board(board)
            count += 1
            if count >= 5:
                if board['7'] == board['8'] == board['9'] == sym: # across the bottom
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['4'] == board['5'] == board['6'] == sym: # across the middle
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['1'] == board['2'] == board['3'] == sym: # across the top
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['1'] == board['4'] == board['7'] == sym: # down the left side
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['2'] == board['5'] == board['8'] == sym: # down the middle
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['3'] == board['6'] == board['9'] == sym: # down the right side
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break 
                elif board['7'] == board['5'] == board['3'] == sym: # diagonal
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                elif board['1'] == board['5'] == board['9'] == sym: # diagonal
                    print("\nGame Over.\n")                
                    print(" **** " +turn + " won. ****")
                    win = False
                    break
                else:
                    turn,sym = player1,choice2.upper()
            else:
                turn,sym = player1,choice2.upper()
    if count == 9 and win == True:
        print("\nGame Over.\n")                
        print("It's a Tie!!")
        
    
if choice == '2':
    print('You have opted for double player mode.')
    player1 = input('Player-1, Enter your name:>>')
    player2 = input('Player-2, Enter your name:>>')
    print(f'{player1} is "X" and {player2}')
    
    count = 0
    turn,sym = player1,'X'
    print_board(board)
    while count != 9:
        print(f'{turn}-{sym}, your turn to play.')
        pos = input(f'{turn},where would you like to play:>>')
        while pos not in board_keys:
            print('invalid entry')
            pos = input('where would you like to play:>>')
        board_keys.remove(pos)
        board[pos] = sym
        print_board(board)
        count += 1
        if count >= 5:
            if board['7'] == board['8'] == board['9'] == sym: # across the bottom
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            elif board['4'] == board['5'] == board['6'] == sym: # across the middle
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            elif board['1'] == board['2'] == board['3'] == sym: # across the top
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            elif board['1'] == board['4'] == board['7'] == sym: # down the left side
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            elif board['2'] == board['5'] == board['8'] == sym: # down the middle
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            elif board['3'] == board['6'] == board['9'] == sym: # down the right side
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win= False
                break 
            elif board['7'] == board['5'] == board['3'] == sym: # diagonal
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            elif board['1'] == board['5'] == board['9'] == sym: # diagonal
                print("\nGame Over.\n")                
                print(" **** " +turn + " won. ****")
                win = False
                break
            else:
                if [turn,sym] == [player1,'X']:
                    turn,sym = player2, 'O'
                else:
                    turn,sym = player1, 'X'
        else:
            if [turn,sym] == [player1,'X']:
                turn,sym = player2, 'O'
            else:
                turn,sym = player1, 'X'                
        
    if count == 9 and win == True:
        print("\nGame Over.\n")                
        print("It's a Tie!!")
```
